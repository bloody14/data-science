```python
import requests # handles HTTP/HTTPS requests
import pandas as pd

def get_api_data():

  reponse = requests.get('https://jsonplaceholder.typicode.com/posts/1')
  data = reponse.json() #convert JSON data into py. dictonary
  print(data)
  df=pd.DataFrame([data]) # this converts dictonary into dataframe
  print(df)
  df.to_csv('data.csv')

get_api_data()
```

```python
# weather appp(multiple cities)
import requests
import pandas as pd

def get_weather_data():

    API_KEY = "2a2070d566cdae6cffea199d2bfc1361"

    cities = ["Jaipur","Delhi", "Mumbai", "Bengaluru"]

    all_weather_data = []

    for city in cities:
        url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"

        response = requests.get(url)
        data = response.json()

        # Check if the API call was successful for the current city
        if data.get("cod") == 200:
            weather_data = {
                "City": data["name"],
                "Temperature": data["main"]["temp"],
                "Feels Like": data["main"]["feels_like"],
                "Humidity": data["main"]["humidity"],
                "Weather": data["weather"][0]["description"],
                "Wind Speed": data["wind"]["speed"]
            }
            all_weather_data.append(weather_data)
        else:
            print(f"Could not retrieve data for {city}: {data.get('message', 'Unknown error')}")

    if all_weather_data:
        df = pd.DataFrame(all_weather_data)
        display(df)
        df.to_csv("weather_data.csv", index=False)
    else:
        print("No weather data was successfully retrieved for any city.")

get_weather_data()
```
