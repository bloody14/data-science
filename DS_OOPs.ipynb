```python
# =========================================
#      COFFEE MACHINE PROJECT (OOP)
# =========================================
# Concepts Used:
# 1. Abstract Class
# 2. Encapsulation
# 3. Inheritance
# 4. Polymorphism
# =========================================

from abc import ABC, abstractmethod


# =========================================
# ABSTRACT CLASS
# =========================================

class Coffee(ABC):

    def __init__(self, name, water, milk, coffee_powder):
        # Encapsulation using private variables
        self.__name = name
        self.__water = water
        self.__milk = milk
        self.__coffee_powder = coffee_powder

    # Getter methods
    def get_name(self):
        return self.__name

    def get_water(self):
        return self.__water

    def get_milk(self):
        return self.__milk

    def get_coffee_powder(self):
        return self.__coffee_powder

    # Abstract Method
    @abstractmethod
    def recipe(self):
        pass

    # Common Method
    def prepare(self):
        print("\nPreparing", self.get_name(), "...\n")
        print("Water :", self.get_water(), "ml")
        print("Milk  :", self.get_milk(), "ml")
        print("Coffee Powder :", self.get_coffee_powder(), "grams")
        self.recipe()


# =========================================
# LATTE CLASS
# =========================================

class Latte(Coffee):

    def __init__(self):
        super().__init__("Latte", 100, 150, 20)

    def recipe(self):

        print("""
        HASHED LATTE ART
        ########################
        ######  LATTE  #########
        ########################
            ( (
             ) )
          ........
          |      |]
          \\      /
           `----'
        """)

        print("Recipe:")
        print("1. Add Espresso Shot")
        print("2. Add Steamed Milk")
        print("3. Add Milk Foam")
        print("4. Latte Ready!")


# =========================================
# AMERICANO CLASS
# =========================================

class Americano(Coffee):

    def __init__(self):
        super().__init__("Americano", 180, 0, 25)

    def recipe(self):

        print("""
        HASHED AMERICANO ART
        ########################
        ####  AMERICANO  ######
        ########################
           __________
          |          |
          |  BLACK   |
          | COFFEE   |
          |__________|
        """)

        print("Recipe:")
        print("1. Add Espresso")
        print("2. Add Hot Water")
        print("3. Americano Ready!")


# =========================================
# BLACK COFFEE CLASS
# =========================================

class BlackCoffee(Coffee):

    def __init__(self):
        super().__init__("Black Coffee", 200, 0, 30)

    def recipe(self):

        print("""
        HASHED BLACK COFFEE ART
        ########################
        #### BLACK COFFEE #####
        ########################
            /////////
           ||||||||||
           || BLACK ||
           ||COFFEE ||
           ||||||||||
        """)

        print("Recipe:")
        print("1. Boil Water")
        print("2. Add Coffee Powder")
        print("3. Stir Well")
        print("4. Black Coffee Ready!")


# =========================================
# MATCHA CLASS
# =========================================

class Matcha(Coffee):

    def __init__(self):
        super().__init__("Matcha", 120, 100, 15)

    def recipe(self):

        print("""
        HASHED MATCHA ART
        ########################
        ###### MATCHA ##########
        ########################
            *********
           * MATCHA *
            *********
             |    |
             |____|
        """)

        print("Recipe:")
        print("1. Add Matcha Powder")
        print("2. Add Hot Water")
        print("3. Add Milk")
        print("4. Matcha Ready!")


# =========================================
# CAPPUCCINO CLASS
# =========================================

class Cappuccino(Coffee):

    def __init__(self):
        super().__init__("Cappuccino", 80, 120, 22)

    def recipe(self):

        print("""
        HASHED CAPPUCCINO ART
        ########################
        ##### CAPPUCCINO ######
        ########################
             ~~~~~~~~
           [ CAPPU ]
           [CCINO ]
            ------
        """)

        print("Recipe:")
        print("1. Add Espresso")
        print("2. Add Steamed Milk")
        print("3. Add Thick Foam")
        print("4. Cappuccino Ready!")


# =========================================
# COFFEE MACHINE
# =========================================

class CoffeeMachine:

    def show_menu(self):

        print("\n========== COFFEE MACHINE ==========")
        print("1. Latte")
        print("2. Americano")
        print("3. Black Coffee")
        print("4. Matcha")
        print("5. Cappuccino")
        print("6. Exit")

    def make_coffee(self, choice):

        if choice == 1:
            coffee = Latte()

        elif choice == 2:
            coffee = Americano()

        elif choice == 3:
            coffee = BlackCoffee()

        elif choice == 4:
            coffee = Matcha()

        elif choice == 5:
            coffee = Cappuccino()

        else:
            print("Invalid Choice")
            return

        coffee.prepare()


# =========================================
# MAIN PROGRAM
# =========================================

machine = CoffeeMachine()

while True:

    machine.show_menu()

    choice = int(input("\nEnter Your Choice : "))

    if choice == 6:
        print("\nThank You For Using Coffee Machine ☕")
        break

    machine.make_coffee(choice)
```
