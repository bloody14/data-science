```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/TSLA.csv")
df
```

```python
df=df.drop(["Date"],axis=1)
df
```

```python
X=df .iloc[:,:5]
Y=df .iloc[:,5]
X
```

```python
Y
```

**Normalization**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
X
```

```python
Y=np.array(Y)
Y
```

```python
Y=Y.reshape(-1,1)
Y
```

```python
# split X and Y into train and test
m=len(X)
X_train=X[:int(0.8*m)]
X_test=X[int(0.8*m):]
Y_train=Y[:int(0.8*m)]
Y_test=Y[int(0.8*m):]

```

```python
len(Y_train)
```

```python
X[0]
```

**initialization**

```python
# Initialization
n=len(X[0])
W=np.random.randn(n,1)
b=0.23

# Model
def LinearRegression(X,W,b):
  return np.dot(X,W)+b

mse=[]
epochs=1000000
learning_rate=0.1
for epoch in range(epochs):
  Y_pred=LinearRegression(X_train,W,b)
  loss=np.mean((Y_pred-Y_train)**2)
  mse.append(loss)
  dW=np.dot(X_train.T,(Y_pred-Y_train))/len(X_train)
  db=np.mean(Y_pred-Y_train)
  #Update
  W=W-learning_rate*dW
  b=b-learning_rate*db

  if(epoch%10 == 0):
    print(f"Epoch : {epoch}  Loss : {loss}")

  if(loss <= 1e-6):
    print("Loss less than 1e-6, Learning completed")
    break






```

```python
import matplotlib.pyplot as plt
plt.plot(mse)
plt.show
```

```python
Y_predicted=LinearRegression(X_train,W,b)
Y_predicted
```
