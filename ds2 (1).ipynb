```python
class bank:
  def __init__(self,name ,acc_no,balance):
    self.name = name
    self.acc_no = acc_no
    self.balance = balance
  def deposit(self,amount):

    self.balance += amount
    return f"Your balance is {self.balance}"
  def withdrwal(self, amount):
    if amount>=self.acc_no:
      self.balance-=amount
    else:
      f"Insufficient balance"
    return f"Your balance is {self.balance}"
cust1=bank("abc",123,1000)
cust1.deposit(500)
cust1.withdrwal(200)



```
